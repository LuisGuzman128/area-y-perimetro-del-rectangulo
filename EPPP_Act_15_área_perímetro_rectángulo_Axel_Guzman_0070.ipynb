* El rectángulo es una figura geométrica que tiene cuatro lados, donde los lados opuestos son iguales. Para poder medir un rectángulo, se utilizan dos conceptos importantes: el área y el perímetro.

* El área de un rectángulo sirve para conocer cuánto espacio ocupa la figura. Se calcula multiplicando la base por la altura. Esto es útil, por ejemplo, cuando queremos saber cuánto mide el espacio de un terreno o una habitación.

* Por otro lado, el perímetro es la suma de todos los lados del rectángulo. Se obtiene sumando dos veces la base y dos veces la altura. El perímetro nos ayuda a saber cuánto mide el contorno de la figura, como cuando queremos poner una cerca alrededor de un terreno.

```python
# programa para calcular area y perimetro de un rectangulo
# funcion para calcular
# formulas del area y perimetro del rectangulo

def arearectangulo():
    base = float(input("Ingresa la base: "))
    altura = float(input("Ingresa la altura: "))
    area = base * altura
    print("El área del rectángulo es:", area)

def perimetrorectangulo():
    base = float(input("Ingresa la base: "))
    altura = float(input("Ingresa la altura: "))
    perimetro = 2 * (base + altura)
    print("El perímetro del rectángulo es:", perimetro)

def menu():
    opcion = 0
    while opcion != 3:
        print("\n--- MENÚ RECTÁNGULO ---")
        print("1. Calcular área del rectángulo")
        print("2. Calcular perímetro del rectángulo")
        print("3. Salir")

        opcion = int(input("Elige una opción: "))

        if opcion == 1:
            arearectangulo()
        elif opcion == 2:
            perimetrorectangulo()
        elif opcion == 3:
            print("Saliendo del programa...")
        else:
            print("Opción no válida")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```
